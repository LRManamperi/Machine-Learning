```python
!pip install pyannote.audio

from pyannote.audio import Model, Inference

# Load the WeSpeaker ResNet34 pretrained model
model = Model.from_pretrained("pyannote/wespeaker-voxceleb-resnet34-LM")

# Set inference mode (window="whole" for full audio at once)
inference = Inference(model, window="whole")
```

```python
embedding1 = inference("spk1_snt1.wav")
embedding2 = inference("spk2_snt4.wav")
```

```python
import numpy as np
from scipy.spatial.distance import cdist

# Ensure 2D shape
embedding1 = np.expand_dims(embedding1, axis=0)
embedding2 = np.expand_dims(embedding2, axis=0)

# Now compute cosine distance
distance = cdist(embedding1, embedding2, metric="cosine")[0, 0]

print(f"Cosine Distance: {distance:.4f}")
if distance < 0.5:
    print("Likely same speaker")
else:
    print("Likely different speakers")
```
